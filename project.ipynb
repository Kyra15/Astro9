```python
# search MAST database for WASP-18b lightcurves (limit amt returned to 1)
search_result = lk.search_lightcurve('WASP-18b', limit=1)
lc = search_result.download()
# convert to csv and plot
lc.to_csv("wasp18blc.csv", overwrite=True)
lc.plot()
```
